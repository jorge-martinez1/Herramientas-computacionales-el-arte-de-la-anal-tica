```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2) Cargar dataset
df = pd.read_csv("games.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
numerical_columns = []
categorical_columns = []
for column in df.columns:
    if df[column].dtype == 'object':
        categorical_columns.append(column)
    else:
        numerical_columns.append(column)
```

```python
preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numerical_columns),
    ("cat", OneHotEncoder(), categorical_columns)
])
```

```python
X = preprocessor.fit_transform(df)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
numerical_columns = []
categorical_columns = []
for column in df.columns:
    if df[column].dtype == 'object':
        categorical_columns.append(column)
    else:
        numerical_columns.append(column)
```

```python
preprocessor = ColumnTransformer([
    ("num",
     Pipeline([
         ("imputer", SimpleImputer(strategy="median")),
         ("scaler", StandardScaler())
     ]),
     numerical_columns),
    ("cat",
     Pipeline([
         ("imputer", SimpleImputer(strategy="most_frequent")),
         ("onehot", OneHotEncoder(handle_unknown="ignore"))
     ]),
     categorical_columns)
])
```

```python
X = preprocessor.fit_transform(df)
```

```python
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
    diff = inertias[-2] - inertias[-1] if k > 1 else 0
    print(f'K={k}, Inercia = {kmeans.inertia_:.6f} â€“ Diferencia = {diff:.6f}')
```

```python
sns.lineplot(x=range(1, 11), y=inertias, marker="o")
plt.xlabel("NÃºmero de clusters (k)")
plt.ylabel("Inercia (SSE - Suma de errores al cuadrado)")
plt.title("MÃ©todo del codo para determinar el nÃºmero Ã³ptimo de clusters")
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
df["Cluster"] = kmeans.labels_
df.head()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df["Cluster"])
plt.title("VisualizaciÃ³n de clusters con PCA")
plt.show()
```
